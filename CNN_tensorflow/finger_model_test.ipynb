```python
from tensorflow.keras.models import load_model

model=load_model('finger_classification.h5')
```

```python
model.summary()
```

```python
import cv2
import matplotlib.image as mpimg
im=cv2.imread('C:/CNN_tensorflow/finger_test.jpg')
im=cv2.resize(im,(150,150))

import matplotlib.pyplot as plt
plt.imshow(im)
plt.axis('Off')
plt.show()
```

```python
import numpy as np
from keras.models import Sequential

categories=['one_finger','two_finger','three_finger','four_finger','five_finger']
data=np.asarray(im)
X = np.array(data)
X = X.astype("float") / 256
X = X.reshape(-1, 150, 150,3)

pred = model.predict(X)
result = [np.argmax(value) for value in pred]
print('New data category : ',categories[result[0]])
```
